```python
#riversBrazil_clipped=gpd.clip(rivers,brazil)
ferrov=gpd.read_file(os.path.join("maps","ferrovias","Red_ferroviaria.shp"))
prov=gpd.read_file(os.path.join("maps","provincias","Provincias.shp"))
```

```python
#brazil_5641=brazil.to_crs(5641)

#rivers_brazil_5641=riversBrazil_clipped.to_crs(brazil_5641.crs)
chile_5361=chile.to_crs(5361)
prov_chile_5361=prov.to_crs(5361)

# Asegurar que ambos estn en el mismo CRS antes del clip
ferrov_aligned = ferrov.to_crs(chile.crs)

# Realizar el clip
ferrovChile_clipped = gpd.clip(ferrov_aligned, chile)

# Luego proyectar al CRS deseado (por ejemplo EPSG:5361)
ferrov_chile_5361 = ferrovChile_clipped.to_crs(5361)
```

```python
chile.plot(facecolor="azure",edgecolor='red',linewidth=0.1)
```

```python
import pandas as pd
infoairports=pd.read_csv(os.path.join("data","chile-airports.csv"))

# some rows

infoairports.iloc[[0,1,2,3,-4,-3,-2,-1],:] #head and tail
```

```python
# bye first row
infoairports.drop(index=0,inplace=True)
infoairports.reset_index(drop=True, inplace=True)
infoairports.head()
```

```python
# keep the  columns needed

infoairports.columns
```

```python
keep=['name','type','latitude_deg', 'longitude_deg','elevation_ft','region_name','municipality']
infoairports=infoairports.loc[:,keep]

infoairports.info()
```

```python
numericCols=['latitude_deg', 'longitude_deg','elevation_ft']
infoairports[numericCols]=infoairports.loc[:,numericCols].apply(lambda x:pd.to_numeric(x))
```

```python
# now
infoairports.info()
```

```python
# let's plot

base = chile_5361.plot(color='white', edgecolor='black') #unprojected

infoairports.plot.scatter(x = 'longitude_deg', y = 'latitude_deg',ax=base)
```

```python
airports=gpd.GeoDataFrame(data=infoairports.copy(),
                 geometry=gpd.points_from_xy(infoairports.longitude_deg,
                                             infoairports.latitude_deg),
                 crs=chile.crs.to_epsg())# the coordinates were in degrees - unprojected

# let's plot

base = chile_5361.plot(color='white', edgecolor='black')
airports.plot(ax=base)
```

```python
#remember:
type(airports), type(infoairports)
```

```python
airports_5361=airports.to_crs(5361)

## then

base = chile_5361.plot(color='white', edgecolor='black')
airports_5361.plot(ax=base)
```

```python
airports_5361['type'].value_counts() # this will not work: airports.type.value_counts()
```

```python
airports_5361.rename(columns={'type':'kind'},inplace=True)

## adding the airports to GPKG
airports_5361.to_file(os.path.join("maps","chileMaps_5361.gpkg"), layer='airports', driver="GPKG")
```
